```python
import pandas as pd
```

```python
data=pd.read_csv('titanic_clean.csv')
```

```python
summary=data.describe()
```

```python
print(summary)
```

```python
print(data.describe(include=['object']))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(data['age'].dropna(),bins=20,edgecolor='black',color='skyblue')
plt.title('Age Distribution')
plt.xlabel('age')
plt.ylabel('frequency')
plt.show()
```

```python
plt.title('Age Distribution')
plt.xlabel('age')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
```

```python
data['pclass'].value_counts().plot(kind='bar',color='lightcoral') # Changed 'Pclass' to 'pclass'
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
```

```python
plt.boxplot(data['age'].dropna(), patch_artist=True,boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()
```

```python
import seaborn as sns
```

```python
sns.pairplot(data[['age', 'fare', 'sibsp', 'parch']].dropna())
plt.show()
```

```python
correlation_matrix = data[['age', 'fare', 'sibsp', 'parch']].corr()
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='survived',hue='survived', data=data, palette='Set2')
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
sns.violinplot(x='pclass', y='age',hue='pclass', data=data, palette='muted')
plt.title('Age Distribution by Passenger Class')
plt.show()
```
